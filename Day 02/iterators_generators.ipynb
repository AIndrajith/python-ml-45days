```python
class EvenNumbers:
    def __init__(self, max_value):
        self.current = 0
        self.max_value = max_value

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.max_value:
            raise StopIteration
        self.current += 2
        return self.current
```

```python
for x in EvenNumbers(10):
    print(x, end=" ")
```

```python
def fibonacci(n):
    current_num = 0
    next_num = 1
    for i in range(1, n+1):
        yield current_num      # pauses function and remembers state
        current_num, next_num = next_num, current_num + next_num  # multiple assignment
```

```python
for num in fibonacci(7):
    print(num, end = " ")
```
