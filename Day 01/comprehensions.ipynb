```python
nums = [1,2,3,4,5,6]
```

```python
odd_cubes = [x ** 3 for x in nums if x % 2 == 1]
print(odd_cubes)
```

```python
words = ["ml", "python", "ai", "engineer"]

filtered = {n: len(n) for n in words if len(n) > 2}
print(filtered)
```

```python
nums2 = [1,2,3,4]

from functools import reduce

factorial = reduce(lambda x,y: x * y, nums2)
print(factorial)
```

```python
labels = [0, 1, 1, 0, 1]

label_dict = {index: label for index, label in enumerate(labels)}
print(label_dict)
```

```python
nums3 = [1,2,2,3,4,5,5,5]

unique = {n for n in nums3}
print(unique)
```

```python
odd = list(filter(lambda x:x%2==1, nums3))
print(odd)
```

```python
squared = list(map(lambda x : x ** 2, nums3))
print(squared)
```

```python
product = reduce(lambda x, y: x*y, nums3)
print(product)
```
