```python
def safe_divide(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        return None
```

```python
result = safe_divide(18, 0)
print(result)
```

```python
def load_path(path):
    try:
        with open(path) as f:
            return f.read()
    except FileNotFoundError as e:
        raise FileNotFoundError(f"Dataset not found at {path}") from e
```

```python
class InvalidDataError(Exception):
    pass

def train_model(X, y):
    if not X or not y:
        raise InvalidDataError("Training data connot be None of Empty")

    print("Training model...")

train_model([],[])
```
